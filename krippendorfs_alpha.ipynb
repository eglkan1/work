```python
files=['ignas.json','goda.json', 'emilijab.json', 'bohdan.json', 'emilijak.json', 'rutak.json']

def merge_JsonFiles(filename):
        result = list()
        for f1 in filename:
            with open(f1, 'r') as infile:
                result.extend(json.load(infile))
    
        with open('all.json', 'w') as output_file:
            json.dump(result, output_file)
    
merge_JsonFiles(files)
```

```python
with open('all.json') as json_file:
    data = json.load(json_file)
```

## Make agreement table for Krippendorff

```python
from collections import Counter
```

```python
merged.iloc[0]['G']
```

```python
merged_row = merged.iloc[0]['G'] + merged.iloc[0]['EK'] + merged.iloc[0]['EB']
```

```python
merged_row
```

```python
Counter(merged_row)
```

```python
Counter(merged_row)[0]
```

```python
agreement_table = []
for i, row in  merged.iterrows():
    merged_row = row['G'] + row['EK'] + row['EB']
    counted = Counter(merged_row)
    agreement_row = []
    for index in range(0,16):
        agreement_row.append(counted[index])
        agreement_table.append(agreement_row)
# print(agreement_table)
```

## Calculate Krippendorff's Alpha

```python
res = kd.alpha(value_counts=agreement_table, level_of_measurement="nominal")
```
